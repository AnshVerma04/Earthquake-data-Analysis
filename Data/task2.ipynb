```python
import pandas as pd
```

```python
read_file = pd.read_csv('Earthquake_data.csv')
read_file.head(5)
```

```python
# check missing values 
read_file.isnull().sum()
```

```python
# Check duplicate data
read_file.duplicated().sum()
```

```python
# Data type
read_file.dtypes
```

```python
# Unique values in categorical columns
read_file['Date'].unique()
```

```python
# Value range check
read_file.describe()
```

```python
# String formatting issues
read_file['Type'].str.strip().unique()
```

```python
# Overall completeness of the dataset
total_cells = read_file.size
missing_cells = read_file.isnull().sum().sum()

completeness_percentage = ((total_cells - missing_cells) / total_cells) * 100
completeness_percentage
```

```python
# column - wise completeness
(1 - read_file.isnull().mean()) * 100
```
